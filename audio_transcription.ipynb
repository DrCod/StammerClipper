```python
import glob
import re

age ={}
sex ={}
stutters ={}

filenames = glob.glob("AllAudio/*.mp3")

def get_age(audio_file):
    file = re.split(r'/',audio_file)
    age = file[1][7:9]
    return age

def get_sex(audio_file):
    
    if not isinstance(audio_file,str):
        print ("Incorrect file format")
    sex =re.split(r'/',audio_file)[1][0]
    
    return sex    
```

```python
# create a dataframe of all audio and their corresponding transcriptions
import pandas as pd
import glob
import re

data = {}

filenames = glob.glob("AllAudio/*.mp3")

def get_transcription(filename):
    sound =AudioSegment.from_mp3(filename)
    sound.export('WavAudio/{}.wav'.format(re.split(r'.',filename)[0]), format='wav')
    r =sr.Recognizer()
    with sr.AudioFile('WavAudio/{}.wav'.format(re.split(r'.',filename)[0])) as src:
        audio =r.record(src)
        try:
            text =r.recognize_google(audio, language='en-US',show_all= True)
        except Exception as e:
            print("Everything is  fine.")
            print(e)
    return text

for f in filenames:
    data[f] = [get_age(f),get_sex(f), get_transcription(f)]
    
```

```python
pd.DataFrame.from_dict(data, orient ='index').to_csv('stutter.csv')
```
